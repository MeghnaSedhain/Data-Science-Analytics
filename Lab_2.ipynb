#Empirical Distribution  for Numerical Data

```python
import numpy as n
import numpy as np
num_data = np.array([12,15,14,10,18,20,22,19,17,16,14,13,11])
```

#Histogram
#Shows the frequency of values grouped into bins; approximates the empirical PDF.

```python
import matplotlib.pyplot as plt
plt.hist(num_data, bins=5, density=True)
plt.title("Histogram - Numerical Data")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

#Empirical CDF
#Description: Shows cummulative probability P(X<x).

```python
x = np.sort(num_data)
y = np.arange(1, len(num_data)+1) / len(num_data)

plt.plot(x,y)
plt.title("ECDF - Numerical Data")
plt.xlabel("Value")
plt.ylabel("Cummulative Probability")
plt.show()
```

```python
from scipy.stats import gaussian_kde

kde = gaussian_kde(num_data)
x_vals = np.linspace(min(num_data), max(num_data), 100)
y_vals = kde(x_vals)

plt.plot(x_vals, y_vals)
plt.title("KDE - Numerical Data")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()

```

#Box Plot
#Shows median, quartiles and outliers

```python
plt.boxplot(num_data)
plt.title("Box Plot")
plt.show()
```

#Violin plot

```python
plt.violinplot(num_data)
plt.title("Violin Plot")
plt.show()
```
